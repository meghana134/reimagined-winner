```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams ['figure.figsize']=(20.0,10.0)

data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head(5)
```

```
#collecting the values of x and y
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
```

```
#mean of x and y
mean_x=np.mean(x)
mean_y=np.mean(y)

#total number of values in x
m=len(x)

#using formula calculate value of b1 and b2
numer=0
denom=0
for i in range(m):
  numer+= (x[i]-mean_x)*(y[i]-mean_y)
  denom+= (x[i]-mean_x) ** 2
  b1=numer/denom
  b0=mean_y - (b1*mean_x)

  #print the values of b1 and b2
  print(b1,b0)
```

```
#plotting values and regression lines
max_x=np.max(x) + 100
min_x=np.min(x) - 100

#calculating the values of x and y
x=np.linspace(min_x,max_x,1000)
y=b0 + b1*x

#plotting line
plt.plot(x,y,color='#58b970',label='regression line')
#plotting scattered plots
plt.scatter(x,y,c='#ef5423',label='scatterplot')

plt.xlabel('head in cm3')
plt.ylabel('brain weights in grams')
plt.legend()
plt.show()

```
